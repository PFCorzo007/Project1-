# Import dependencies 

```python
import matplotlib.pyplot as plt 
import pandas as pd
```

# Capture dataset with Pandas

```python
data = {'Year': [2013,2014,2015,2016,2017,2018,2019,2020,2021],
        'Number of Crypto': [66,506,562,644,1335,1604,2816, 4501,6826]
}
```

```python
df = pd.DataFrame (data)
print(df)
```

# Use pyplot to create line graph 

```python
plt.plot(df['Year'], df['Number of Crypto'], color='blue', marker='o')
plt.title('Crypto Exponential Growth')
plt.xlabel('Year',fontsize=12)
plt.ylabel('New Cryptocurrencies Launched', fontsize=12)
plt.grid(True)
plt.show()
```
